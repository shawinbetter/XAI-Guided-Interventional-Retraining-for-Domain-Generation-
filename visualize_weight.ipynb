```python
from tensorflow.keras.models import load_model
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
model = load_model("BaselineModel.hdf5")
w,b = model.layers[-1].get_weights()
```

```python
nw,nb = np.load("GradCamTraining_0.3_0.5_20_0.3.npy",allow_pickle=True)
```

```python
plt.figure(figsize = [8,3],dpi=200)
plt.xlabel('weights')
plt.ylabel('counts')
plt.hist(w [abs(w) <= 0.2],bins = 50,label = 'pretrained',alpha = 0.2);
plt.hist(nw [abs(nw) <= 0.2],bins = 50,label = 'XGIR',alpha = 0.2);
plt.legend()
```

```python
plt.figure(figsize = [8,3],dpi=200)
plt.xlabel('bias')
plt.ylabel('counts')
plt.hist(b ,bins = 50,label = 'pretrained',alpha = 0.2);
plt.hist(nb ,bins = 50,label = 'XGIR',alpha = 0.2);
plt.legend()
```

```python
plt.figure(figsize = [6,4], dpi= 150)
sns.heatmap(w,vmax=np.max(nw),vmin = np.min(nw))
```

```python
plt.figure(figsize = [6,4], dpi= 150)
sns.heatmap(nw,vmax=np.max(nw),vmin = np.min(nw))
```

```python
plt.figure(figsize = [6,4], dpi= 150)
plt.hist((nw - w).reshape(2048*1000,1),bins = 100);
```
