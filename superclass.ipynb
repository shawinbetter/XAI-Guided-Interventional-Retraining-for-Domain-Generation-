```python
import pandas as pd
import tensorflow as tf
import numpy as np
import json
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import matplotlib.image as img
import seaborn as sns
import os
```

```python
car_names = ['ambulance','beach_wagon','cab','convertible','jeep',
            'limousine','minivan','Model_T','racer','sports_car']

class_labels = json.load(open('CWOX-1-main/CWOX/imagenet_class_index.json', 'r'))

# folder : nums_label
folder_to_nums = {class_labels[key][0]: key for key in class_labels.keys()}

# class name : folder
class_folder = {v[1]:v[0] for k,v in class_labels.items() if v[1] in car_names}
```

```python
print(' '.join(list(class_folder.values())))
```

#### Visualize

```python
plt.figure(figsize = [13,5.5], dpi = 200)
for i in range(10):
    plt.subplot(2,5,i+1)
    folder = list(class_folder.items())[i][1]
    path = 'val/' + folder + "/" + os.listdir('val/' + folder + "/")[0]
    img = tf.keras.preprocessing.image.load_img(path, target_size=(299,299))
    plt.imshow(img)
    plt.title(list(class_folder.items())[i][0])
plt.savefig('cars.png',dpi = 200)
```
